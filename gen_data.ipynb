```python
import csv
from openai import OpenAI
client = OpenAI()
results=[]

with open('prompts3_multilabels.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Label', 'Prompt', 'Generated Counter Argument']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    for lab, text in zip(label, texts):
        # Create a prompt with the text
        # if there are more than one labels then only do else continue

        prompt = f"Give a short counterpoint to the statement in this anti-vaccine tweet:\n{text}\n\n"

        # print(prompt)
        # Generate the completion
        completion = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[{"role": "user", "content": prompt}]
        )
        generated_text = completion.choices[0].message.content
        # print(generated_text)
        # Write the result to the CSV file
        writer.writerow({
            'Label': lab,
            'Prompt': prompt,
            'Generated Counter Argument': generated_text
        })

        # Also store in the list for future use if needed
        results.append(generated_text)

print("All results have been saved to 'counter_arguments_prompt3.csv'.")
```

```python
test_data=pd.read_csv('/home/sohampoddar/HDD2/utsav/Dataset/test.csv')
data = test_data.sample(n=500, random_state=1)
# print how many rows with multiple labels
multiple_labels = data[data['labels'].str.split().apply(len) > 1]
single_labels = data[data['labels'].str.split().apply(len)==1]
```

```python
len(multiple_labels), len(single_labels)
```

```python
all_labels = ['religious','political','ingredients','unnecessary','conspiracy','mandatory','ineffective','side-effect','pharma','rushed','country']

from skmultilearn.model_selection import IterativeStratification

# Convert labels to binary matrix
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
y_multi = mlb.fit_transform(multiple_labels['labels'])
stratifier = IterativeStratification(n_splits=2, order=1, sample_distribution_per_fold=[0.625, 0.375])
test_indices, _ = next(stratifier.split(multiple_labels.index, y_multi))
selected_multi = multiple_labels.iloc[test_indices[:50]]

# Step 4: Stratified sampling for single-labeled tweets
# Ensure all labels are represented
# Count occurrences of each label in single-label tweets
label_counts = single_labels['labels'].explode().value_counts()

# Calculate the minimum number of tweets per label
min_tweets_per_label = 30 // len(all_labels)
selected_single = pd.DataFrame()
for label in all_labels:
    label_tweets = single_labels[single_labels['labels'].apply(lambda x: label in x)]
    sample_size = min(min_tweets_per_label, len(label_tweets))
    selected_single = pd.concat([selected_single, label_tweets.sample(sample_size, random_state=42)])
# If we still need more tweets, sample randomly
remaining_tweets = 30 - len(selected_single)
if remaining_tweets > 0:
    additional_tweets = single_labels[~single_labels.index.isin(selected_single.index)].sample(remaining_tweets, random_state=42)
    selected_single = pd.concat([selected_single, additional_tweets])
```

```python
# Step 5: Combine the subsets
final_dataset = pd.concat([selected_multi, selected_single])

# Step 6: Verify label distribution
print("Final dataset label distribution:")
print(final_dataset['labels'].explode().value_counts())

# Save or use the final dataset
final_dataset.to_csv('selected_tweets.csv', index=False)
```

```python
import csv
from openai import OpenAI
client = OpenAI()
with open('counter_argument_80_with_label.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Label', 'Prompt', 'Generated Counter Argument']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    for lab, text in zip(label, texts):
            # if there are more than one labels then only do else continue
            prompt = f"Generate a brief counter argument for the anti-vaccine tweet you may also include some relevant links to support your argument:\n{text}\nTalk About "
            lab = lab.split()
            # print(lab)
            if isinstance(lab, list):
                # print("List")
                mapped_labels = " and ".join([labels_map.get(l, "") for l in lab])  # Use get to handle potential missing keys
            else:
                # print('Not list')
                mapped_labels = labels_map.get(lab, "")  # Handle the case where lab is a single string
            prompt += mapped_labels
            # print("Prompt is:",prompt,"\n")
            # Generate the completion
            completion = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {
                    "role": "user",
                    "content": prompt
                }
            ]
            )
            generated_text = completion.choices[0].message.content

            # Write the result to the CSV file
            writer.writerow({
                'Label': lab,
                'Prompt': prompt,
                'Generated Counter Argument': generated_text
            })
```
